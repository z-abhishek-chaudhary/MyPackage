```python
import pandas as pd
df = pd.read_csv('responses.csv', low_memory=False)
df['year'] = pd.DatetimeIndex(df['created_time']).year
```

```python
table = df.groupby(['year', 'brand_name', 'user_id', 'story_id', 'node_key']).size().reset_index(name='count')
```

```python
table1 = df.groupby(['year']).agg({
    'brand_name': 'nunique',
    'user_id': 'nunique',
    'story_id': 'nunique',
    'video_id': 'nunique'
    }
)

table1
```

```python
table2 = df.groupby(['year','brand_name']).agg({
    'user_id': 'nunique',
    'story_id': 'nunique',
    'video_id': 'nunique'
    }
)
table2
```

```python
table3 = df.groupby(['year','brand_name','user_id']).agg({
    'story_id': 'nunique',
    'video_id': 'nunique'
    }
)
table3
```

```python
table4 = df.groupby(['year','brand_name','user_id','story_id']).agg({
    'video_id': 'nunique'
    }
)
table4
```
